```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from GUIs import rb_spec as r 
from IGM import rb_setline as setline
targets=['SDSSJ155304.92+354828.6','SDSSJ140505.77+470441.1','SDSSJ124035.51+094941.0','HS1102+3441','CSO-0873','J112632.91+120437.3','SDSSJ134251.60-005345.3']
zabs=[.4754,.357,0.37097,.28925,0.66042,0.51992,0.22722]
transition= 1031.2

i=3
filename=targets[i]+'_target/'+targets[i]+'_coadd_FUVM_final_lpALL.fits.gz'
if i==1:
    filename=targets[i]+'_target/SDSSJ140505.77+470441.1_coadd_G140L_final_lpALL.fits'
    
print(filename)
```

```python

s=r.read_spec(filename,filetype='HSLA')
s.shift_spec(zabs[i]);
xlim=[-600,600]
s.slice_spec(transition,xlim[0],xlim[1],use_vel=True);
str1=setline.rb_setline(transition,'closest',linelist='LLS')
s.fit_continuum(mask=[-100,100],domain=xlim,Legendre=3)
plt.subplot(2,1,1)
plt.step(s.velo,s.flux_slice)
plt.step(s.velo,s.flux_slice/s.fnorm)
plt.xlim(xlim)
plt.subplot(2,1,2)
plt.step(s.velo,s.fnorm)
plt.plot([-600,600],[1,1],'--')
plt.xlim(xlim)

plt.show()
s.compute_EW(transition,vmin=-200.,vmax=300.);
#s.save_slice(outfile_spec)
```
